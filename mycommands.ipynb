```python
#1 sort combined score and put 10 hightest and 10 lowest in file
import pandas as pd
import numpy as np

scores_raw = pd.read_csv("scores.txt", sep='\t', header=None)
scores = scores_raw.copy()
```

```python
#9 running list of best scores
scores = scores_raw.copy()
best = None

for chunk in pd.read_csv("scores.txt", sep="\t", header=None, chunksize=2_500):
    score_cols = chunk.columns[1:]
    chunk["combined_score"] = chunk[score_cols].sum(axis=1)
    if best is None:
        cand = chunk # Fixed your error baby
    else:
        cand = pd.concat([best, chunk])
    best = cand.nlargest(10, "combined_score")
    
best = best.sort_values(by="combined_score", ascending=False).drop(columns="combined_score")
best.to_csv(
    "scoresextreme9.txt",
    sep="\t",
    header=False,
    index=False
)
```
